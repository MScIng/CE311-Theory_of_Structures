## Problem 1 : 

For the given beam, find the support reactions on point A. 

```python
# Input values 
F = 10 # kN
xf = 1 #m
L = 10 #m
```

```python
def A_vertical( F , L , xf ) :
    return( F * ( L - xf) / L ) 
```

```python
Ay = A_vertical( F , L ,xf )
print( f"Vertical support reaction at point A is {Ay} kN")
```

## Problem 2

For the fiven system, obtain M & V diagrams?

```python
def moment( x , xf , F , Ay , L ): 
    if x <= xf : 
        temp_moment = Ay * x 
    elif x > xf and x <= L : 
        temp_moment = Ay * x - F * ( x - xf) 
    # 
    return( temp_moment ) 
        
```

```python
moment(  L/2  , xf , F , Ay , L ) 
```

```python
# Define the location of the Constant value anywhere between 0 and L
xf_any = range( 0 , 11)
#[ print( item ) for item in xf_any ] 
A_vertical_list = []
for xf in xf_any : 
    dummy_A_vertical = A_vertical( F , L ,xf)
    #print( dummy_A_vertical)
    A_vertical_list.append( dummy_A_vertical) 
```
